```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(docs[0].page_content, "html.parser")  # HTML 파서 초기화
content = soup.find_all("div")  # 모든 div 태그 검색
```

```python
import re

cur_fs = None
cur_text = ""
snippets = []  # 동일한 글꼴 크기의 모든 스니펫 수집
for c in content:
    sp = c.find("span")
    if not sp:
        continue
    st = sp.get("style")
    if not st:
        continue
    fs = re.findall("font-size:(\d+)px", st)
    if not fs:
        continue
    fs = int(fs[0])
    if not cur_fs:
        cur_fs = fs
    if fs == cur_fs:
        cur_text += c.text
    else:
        snippets.append((cur_text, cur_fs))
        cur_fs = fs
        cur_text = c.text
snippets.append((cur_text, cur_fs))
# 중복 스니펫 제거 전략 추가 가능성 (PDF의 헤더/푸터가 여러 페이지에 걸쳐 나타나므로 중복 발견 시 중복 정보로 간주 가능)
```

```python
from langchain_core.documents import Document

cur_idx = -1
semantic_snippets = []
# 제목 가정: 높은 글꼴 크기
for s in snippets:
    # 새 제목 판별: 현재 스니펫 글꼴 > 이전 제목 글꼴
    if (
        not semantic_snippets
        or s[1] > semantic_snippets[cur_idx].metadata["heading_font"]
    ):
        metadata = {"heading": s[0], "content_font": 0, "heading_font": s[1]}
        metadata.update(docs[0].metadata)
        semantic_snippets.append(Document(page_content="", metadata=metadata))
        cur_idx += 1
        continue

    # 동일 섹션 내용 판별: 현재 스니펫 글꼴 <= 이전 내용 글꼴
    if (
        not semantic_snippets[cur_idx].metadata["content_font"]
        or s[1] <= semantic_snippets[cur_idx].metadata["content_font"]
    ):
        semantic_snippets[cur_idx].page_content += s[0]
        semantic_snippets[cur_idx].metadata["content_font"] = max(
            s[1], semantic_snippets[cur_idx].metadata["content_font"]
        )
        continue

    # 새 섹션 생성 조건: 현재 스니펫 글꼴 > 이전 내용 글꼴, 이전 제목 글꼴 미만
    metadata = {"heading": s[0], "content_font": 0, "heading_font": s[1]}
    metadata.update(docs[0].metadata)
    semantic_snippets.append(Document(page_content="", metadata=metadata))
    cur_idx += 1
```

```python
import pdfplumber
from langchain_core.documents import Document  # ✅ LangChain Document 객체

docs = []

with pdfplumber.open(FILE_PATH) as pdf:
    for i, page in enumerate(pdf.pages):
        words = page.extract_words(use_text_flow=True)  # 단어별 좌표 추출
        left_col = []
        right_col = []
        
        # 문서 전체 평균 x 좌표 계산 (기준선을 찾기 위해)
        avg_x = sum(float(w['x0']) for w in words) / len(words)
        
        for word in words:
            if float(word['x0']) < avg_x:
                left_col.append(word['text'])
            else:
                right_col.append(word['text'])

        # ✅ Document 객체로 변환하여 저장
        doc = Document(
            page_content=f"=== Left Column ===\n{' '.join(left_col)}\n\n=== Right Column ===\n{' '.join(right_col)}",
            metadata={"page_number": i + 1}
        )
        docs.append(doc)

# ✅ Document 객체 출력
for doc in docs:
    print(doc.page_content)
```

### Multi modal

```python
import fitz  # PyMuPDF
from PIL import Image
import io

def extract_text_images_from_columns(pdf_path):
    doc = fitz.open(pdf_path)
    extracted_data = []

    for page in doc:
        blocks = page.get_text("blocks")  # 블록 단위로 텍스트 추출
        blocks = sorted(blocks, key=lambda x: (x[1], x[0]))  # (y좌표, x좌표) 기준 정렬
        
        left_column = []
        right_column = []

        page_width = page.rect.width
        column_threshold = page_width / 2  # 중간 지점을 기준으로 왼쪽/오른쪽 컬럼 구분

        for block in blocks:
            x0, y0, x1, y1, text, *_ = block
            if x0 < column_threshold:  # 왼쪽 컬럼
                left_column.append((y0, text))
            else:  # 오른쪽 컬럼
                right_column.append((y0, text))

        # y좌표 기준으로 정렬 후 결합
        left_column = [text for _, text in sorted(left_column)]
        right_column = [text for _, text in sorted(right_column)]
        
        # 이미지 추출
        images = []
        for img_index, img in enumerate(page.get_images(full=True)):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_bytes = base_image["image"]
            img = Image.open(io.BytesIO(image_bytes))
            images.append(img)

        extracted_data.append({
            "left_column": "\n".join(left_column),
            "right_column": "\n".join(right_column),
            "images": images
        })

    return extracted_data

# 실행 예시
pdf_path = "./dataset/artworks_10_6-8.pdf"
data = extract_text_images_from_columns(pdf_path)

# 텍스트 확인
print("Left Column Text:\n", data[0]["left_column"])
print("Right Column Text:\n", data[0]["right_column"])

# 이미지 저장 예시
for idx, img in enumerate(data[0]["images"]):
    img.save(f"/content/image_{idx}.png")
```

```python
import pdfplumber

def extract_text_tables(pdf_path):
    extracted_data = []

    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            text = page.extract_text(layout=True)  # 레이아웃 유지
            tables = page.extract_tables()  # 표 추출
            
            extracted_data.append({
                "text": text,
                "tables": tables
            })
    
    return extracted_data

# 실행 예시
pdf_path = "./dataset/artworks_10_6-8.pdf"
data = extract_text_tables(pdf_path)

# 텍스트 및 표 확인
print("Extracted Text:\n", data[0]["text"])
print("Extracted Tables:\n", data[0]["tables"])
```
